# 颜色生成器

```python
def generate_lighter_blues(base_color, num_colors):
    # 解析基础颜色
    base_rgb = {
        'red': int(base_color[1:3], 16),
        'green': int(base_color[3:5], 16),
        'blue': int(base_color[5:7], 16)
    }
    
    # 计算每次增加的步长
    step = {'red': 5, 'green': 5, 'blue': 0}  # 蓝色分量保持不变
    lighter_blues = [base_color]  # 基础颜色作为第一组颜色

    for _ in range(1, num_colors):
        # 计算下一组颜色的RGB值
        next_rgb = {
            'red': min(base_rgb['red'] + step['red'], 255),
            'green': min(base_rgb['green'] + step['green'], 255),
            'blue': base_rgb['blue']  # 蓝色分量不变
        }
        
        # 将RGB值转换回十六进制格式
        next_color = '#{:02X}{:02X}{:02X}'.format(
            next_rgb['red'],
            next_rgb['green'],
            next_rgb['blue']
        )
        lighter_blues.append(next_color)
        
        # 更新基础颜色为当前颜色
        base_rgb = next_rgb

    return lighter_blues
```

```python
# 使用函数生成50组颜色
base_color = '#4A90E2'  # 最深的蓝色
lighter_blues = generate_lighter_blues(base_color, 35)
print(lighter_blues)
```

```python
def generate_lighter_reds(base_color, num_colors):
    # 解析基础颜色
    base_rgb = {
        'red': int(base_color[1:3], 16),
        'green': int(base_color[3:5], 16),
        'blue': int(base_color[5:7], 16)
    }
    
    # 计算每次增加的步长
    step = {'red': 0, 'green': 5, 'blue': 5}  # 红色分量保持不变
    lighter_reds = [base_color]  # 基础颜色作为第一组颜色

    for _ in range(1, num_colors):
        # 计算下一组颜色的RGB值
        next_rgb = {
            'red': min(base_rgb['red'] + step['red'], 255),
            'green': min(base_rgb['green'] + step['green'], 255),
            'blue': min(base_rgb['blue'] + step['blue'], 255)
        }
        
        # 将RGB值转换回十六进制格式
        next_color = '#{:02X}{:02X}{:02X}'.format(
            next_rgb['red'],
            next_rgb['green'],
            next_rgb['blue']
        )
        lighter_reds.append(next_color)
        
        # 更新基础颜色为当前颜色
        base_rgb = next_rgb

    return lighter_reds

```

```python
# 使用函数生成50组颜色
base_color = '#FF6347'  # 最深的红色
lighter_reds = generate_lighter_reds(base_color, 35)
print(lighter_reds)
```
