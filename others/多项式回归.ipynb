```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # 特征数据，需要是二维数组
y = np.array([2, 4, 6, 8, 10])  # 目标数据
```

```python
degree = 2  # 多项式的度数
poly_features = PolynomialFeatures(degree=degree)
x_poly = poly_features.fit_transform(x)
```

```python
model = LinearRegression()
model.fit(x_poly, y)
```

```python
y_pred = model.predict(x_poly)
mse = mean_squared_error(y, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
x_new = np.array([x[0]])  # 将x[0]转换为二维数组
x_new_poly = poly_features.transform(x_new)  # 使用之前拟合的poly_features转换x_new
y_pred_new = model.predict(x_new_poly)
print(f'预测结果: {y_pred_new[0]}')
```

```python
plt.scatter(x, y, color='red')  # 原始数据点
plt.plot(x, y_pred, color='blue')  # 多项式拟合曲线
plt.show()
```
