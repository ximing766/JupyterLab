```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.lines import Line2D

# 初始化画布
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xlim(-20, 300)
ax.set_ylim(-20, 120)
ax.axis('off')

# 绘制主从门
master_gate = patches.Rectangle((-10, 48), 120, 48, facecolor='gray')
slave_gate = patches.Rectangle((-10, -48), 120, 48, facecolor='gray')
ax.add_patch(master_gate)
ax.add_patch(slave_gate)
ax.text(50, 90, 'Master Gate', ha='center', va='center', fontsize=14, color='white')
ax.text(50, -90, 'Slave Gate', ha='center', va='center', fontsize=14, color='white')

# 定义节点数据（简化示例，按行排列）
nodes_data = [
    # x坐标, y坐标, 颜色, 上半部分文字, 下半部分文字, dBm值
    (140, 48, 'green', '±2.5cm', '±5.6cm', '-51dBm'),
    (140, 0, 'yellow', '±4.9cm', '±2.2cm', '-58dBm'),
    (140, -48, 'red', '±5.2cm', '±28.8cm', '-78dBm'),
    # 更多节点数据...
]

# 绘制节点
for x, y, color, text1, text2, dbm in nodes_data:
    circle = patches.Circle((x, y), 10, facecolor=color, edgecolor='black')
    ax.add_patch(circle)
    ax.text(x, y+5, text1, ha='center', va='center', fontsize=8)
    ax.text(x, y-5, text2, ha='center', va='center', fontsize=8)
    ax.text(x, y+15, dbm, ha='center', va='center', fontsize=8)

# 绘制虚线连接
dashed_line = Line2D([140, 280], [48, 48], linestyle='--', color='gray')
ax.add_line(dashed_line)

# 绘制距离标注
distances = [40, 10, 50, 50, 50, 50]
x_pos = 0
for i, dist in enumerate(distances):
    ax.text(x_pos + dist/2, -120, f"{dist}cm", ha='center', va='center')
    x_pos += dist

plt.show()
```
