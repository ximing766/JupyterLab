```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False     # 

PHONECALL_HEIGHT = 50
y_original = np.arange(0, 101, 1)
y_new = np.sqrt(np.abs(y_original**2 - PHONECALL_HEIGHT**2))

plt.figure(figsize=(10, 6))
plt.plot(y_original, y_new, 'b-', label='y')
plt.plot(y_original, y_original, 'r--', label='y')
plt.xlabel('y (cm)')
plt.ylabel('y (cm)')
plt.title(f'y = sqrt(abs(y$^2$ - {PHONECALL_HEIGHT}$^2$)) ')
plt.grid(True)
plt.legend()

plt.xlim(left=0)
plt.ylim(bottom=0)
plt.axhspan(ymin=0, ymax=50, color='purple', alpha=0.3)  

plt.show()
```
